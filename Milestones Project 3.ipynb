```python
#import libraries

import requests
from bs4 import BeautifulSoup

```

```python
#the website to scrape

url ="https://www.channelnewsasia.com/commentary/malaysia-jobless-unemployment-youth-young-people-graduates-political-stability-4169011"
```

```python
#send a requesBt to the website to retrieve the html content

html= requests.get(url)
```

```python
#use BeautifilSoup to parse the html

bs_cna=BeautifulSoup(html.content, "html.parser")
bs_cna
```

```python
#to extract the content of the news article and only print out text content

body_contents=bs_cna.find_all("div",{"class" :"text-long"})

def extract_text_from_web (body_contents):
    text_content=" "
    for body_content in body_contents:
        text_content += body_content.get_text()
    return text_content

text_content = extract_text_from_web(body_contents)
print(text_content)



```

```python
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')

nltk.download("stopwords")
from nltk.corpus import stopwords
```

```python
#clean the text file from all the punctuation and digits)

from stopwordz import clean_text
import re
text_without_punctuation =clean_text(text_content)

def filter_text(text_without_punctuation):
    text_without_punctuation=re.sub(r'[\d]'," ",text_without_punctuation)
    text_without_punctuation=re.sub("malaysia", " " ,text_without_punctuation)
    text_without_punctuation=re.sub("cent", " ",text_without_punctuation)
    text_without_punctuation= re.sub("malaysian", " " ,text_without_punctuation)
    text_without_punctuation=re.sub("year"," " ,text_without_punctuation)
    return text_without_punctuation

filtered = filter_text(text_without_punctuation)
filtered
```

```python
#tokenize the news articles into word by word list, filter out the stopword

news_word = word_tokenize(filtered)
news_word

list=[]
for word in news_word:
    list.append(word)

list

```

```python
#convet list of strings into single variable and create wordcloud 
string_of_words=" ".join(list)
wordcloud = WordCloud(width = 800, height = 300,background_color ='white', min_font_size = 10)
wordcloud.generate(string_of_words)

```
